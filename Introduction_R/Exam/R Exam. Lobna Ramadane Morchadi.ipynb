```r
dGDP <-  read.csv("GDP.csv")
head(dGDP)
str(dGDP)
```

```r
dHDI <-  read.csv("HDI.csv")
head(dHDI)
str(dHDI)
```

```r
"Sort by province order"
ord_GDP <- (dGDP[order(dGDP$Province), ])
ord_HDI <- (dHDI[order(dHDI$Province), ])
ord_GDP
ord_HDI
```

```r
df_HDI <- subset (ord_HDI, select=-Province)
df_HDI

all_data <- cbind(ord_GDP,df_HDI)
all_data
```

```r
head(all_data) 

tail(all_data)
```

```r
library(ggplot2)
ggplot(all_data, aes(x = GDP.per.capita, y = HDI, color=CCAA)) + geom_point()
```

```r
GDP_m <- aggregate(x = all_data$GDP.per.capita, by = list(all_data$CCAA), FUN = mean)
GDP_m
HDI_m <- aggregate(x = all_data$HDI, by = list(all_data$CCAA), FUN = mean)
HDI_m
```

```r
GDP_sd <- aggregate(x = all_data$GDP.per.capita, by = list(all_data$CCAA), FUN= sd)
GDP_sd
HDI_sd <- aggregate(x = all_data$HDI, by = list(all_data$CCAA), FUN= sd)
HDI_sd
```

```r
GDP_sdv <- GDP_sd
GDP_sdv[is.na(GDP_sdv)] <- 0
GDP_sdv

HDI_sdv <- HDI_sd
HDI_sdv[is.na(HDI_sdv)] <- 0
HDI_sdv
```
